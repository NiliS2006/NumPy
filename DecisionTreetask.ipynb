```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import tree
```

```python
dtype = [
    ('Drama_Level', float),  # Specify data type for Drama_Level
    ('Has_Instagram', int),  # Specify data type for Has_Instagram
    ('Species', 'U20'),  # Specify data type for Species (string, max 20 chars)
    ('Snack_Preference', 'U20'),  # Specify data type for Snack_Preference (string, max 20 chars)
    ('Activity_Score', int),  # Specify data type for Activity_Score
    ('Daily_Hair_Shedding', float),  # Specify data type for Daily_Hair_Shedding
    ('Is_Evil', int)  # Specify data type for Is_Evil
]
data = np.array([
    (8.7, 1, 'Dog', 'Cheese', 6, 15.0, 0),  # Example data
    (9.9, 1, 'Cat', 'Tuna', 7, 5.0, 1),
    (6.5, 0, 'Iguana', 'Souls', 5, 0.1, 0),
    (7.4, 1, 'Dog', 'Tuna', 9, 12.0, 0),
    (10.0, 1, 'Cat', 'Souls', 10, 6.0, 1),
    (9.2, 1, 'Cat', 'Cheese', 8, 4.5, 0),
    (5.5, 0, 'Snake', 'Souls', 4, 0.0, 1),
], dtype=dtype)
```

```python
df = pd.DataFrame(data)
df
```

```python
inputs = df.drop('Is_Evil', axis='columns')
target = df['Is_Evil']
```

```python
inputs
```

```python
target
```

```python
le_Species = LabelEncoder()
le_Snack_Preference = LabelEncoder()
```

```python
inputs['Species_n'] = le_Species.fit_transform(inputs['Species'])
inputs['Snack_Preference_n'] = le_Species.fit_transform(inputs['Snack_Preference'])
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['Species', 'Drama_Level', 'Snack_Preference','Activity_Score','Daily_Hair_Shedding'], axis='columns')
```

```python
inputs_n
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[1,	1,	0,	5]])
```
